```python
from math import factorial
def combinations(n,k):
    return factorial(n)/(factorial(n-k)* factorial(k))
```

## Задание 1  
> Из колоды в 52 карты извлекаются случайным образом 4 карты.  
a) Найти вероятность того, что все карты – крести.  
b) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.
### Решение
a) Вероятность определяется отношением числом исходов, благоприятствующих событию, к общему числу возможных исходов.  
Общее число возможных исходов равно числу сочетаний из 52 по 4 - $ C_{52}^4 $. Число благоприятных исходов равно числу возможностей выбрать 4 карты из 13 крестей - $ C_{13}^4 $ =>  
$P = \frac{C_{13}^4}{C_{52}^4} = \frac{\frac{13!}{9!4!}}{\frac{52!}{48!4!}} = \frac{10*11*12*13}{49*50*51*52} \approx  0.00264$  
б) Полная вероятность равна 1. Поэтому можно посчитать вероятность того, что среди 4 карт не окажется ни одного туза и вычесть ее из 1. Число случаев, при которых среди четырех карт нет ни одного туза равно числу сочетаний из 48 по 4 (48 не тузов) =>
$P = 1 - \frac{C_{48}^4}{C_{52}^4} = 1- \frac{45*46*47*48}{49*50*51*52} =1- 0.7187367254594146 \approx 0.281$



```python
print(f'Точный ответ к заданию a) => {combinations(13,4)/combinations(52,4)}')
print(f'Точный ответ к заданию a) => {1- combinations(48,4)/combinations(52,4)}')
```

## Задание 3
> В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?  
### Решение
 Вероятность определяется отношением числом исходов, благоприятствующих событию, к общему числу возможных исходов.  
Общее число возможных исходов равно числу сочетаний из 15 по 3 - $ C_{15}^3 $. Число благоприятных исходов равно числу сочетаний из 9 по 3 - $ C_{9}^3 $ =>  
$P = \frac{C_{9}^3}{C_{15}^3} = \frac{7*8*9}{13*14*15} \approx  0.1846$  

```python
print(f'Точный ответ к заданию 3 => {combinations(9,3)/combinations(15,3)}')
```

## Задание 4
> В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?
### Решение
Всего существует $C_{100}^2$ возможных сочетаний. Правильным является только одно =>  
P = $\frac{1}{C_{100}^2} = \frac{1}{\frac{100!}{98!2!}} = \frac{98!2!}{100!} = \frac{1*2}{99*100} \approx 0.0002$

```python
print(f'Точный ответ к заданию 4 => {1/combinations(100,2)}')
```

б) вторым спортсменом

где $P(A|B_1) =  0.9$ - вроятность попадания первым спортсменом, $P(A)$ - вероятность попадания.  
Вероятности произведения выстрела каждым из сортменов одинаковы и равны $\frac{1}{3} $. Тогда  
$ P(B_1|A) =  \frac{0.9\frac{1}{3}}{0.9\frac{1}{3} + 0.8\frac{1}{3} + 0.6\frac{1}{3}} = \frac{9}{23}$  
б) По аналогии с вариантом а) можно записать  
$ P(B_2|A) =  \frac{0.8\frac{1}{3}}{0.9\frac{1}{3} + 0.8\frac{1}{3} + 0.6\frac{1}{3}} = \frac{8}{23}$  
в)$P(B_3|A) =  \frac{0.6\frac{1}{3}}{0.9\frac{1}{3} + 0.8\frac{1}{3} + 0.6\frac{1}{3}} = \frac{6}{23}$
