```python
from math import factorial
from  math import exp
def combinations(n,k):
    return factorial(n)/(factorial(n-k)* factorial(k))
def poison(l,m):
    return (exp(-l)*(l**m))/factorial(m)
```

## Задание 1.  
> Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. 
Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.  
### Решение
Вероятность наступления события при одиночном испытании значительно больше, чем 0.1, а число испытаний не превышает 100. В данном случае определение числа произошедших событий (k) при n независимых испытаниях и известной вероятности наступаления события в одном испытании (p) дается формулой Бернулли, описывающей Биномиальное распределение: $P(k=85)= C_{100}^{85}p^kq^{n-k} = \frac{100!}{85!(100-85)!}0.8^{85}0.2^{15} \approx 0.048 $

```python
print(f'Точный ответ к заданию 1 => {combinations(100,85) * 0.8**85 * 0.2**15}')
```

## Задание 2  
> Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек.  
a) Какова вероятность, что ни одна из них не перегорит в первый день?  
b)Какова вероятность что перегорят ровно две?  
### Решение  
Вероятность наступления события при одиночном испытании намного меньше, чем 0.1, а число испытаний не превышает 100. В данном случае определение числа произошедших событий (m)  дается формулой Пуассона: $P_m= \frac{\lambda^m}{m!} e^{-\lambda} $. В этой формуле $\lambda = np = 0.0004 * 5000 = 2$ - число событий в единицу времени. =>  
а) $P_m= \frac{2^0}{0!} e^{-2} = e^{-2} \approx 0.135$  
б)$P_m= \frac{2^2}{2!} e^{-2} = e^{-2} \approx 0.2171$


```python
print(f'Точный ответ к заданию 2.а => {poison(2,0)}')
print(f'Точный ответ к заданию 2.б => {poison(2,2)}')
```

## Задание 3
>  Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?  
### Решение
Вероятность выпадания орла в одном испытании p = 0.5, что значительно превышает 0.1. Перед нами классческое биномиальное распределение, описываемое формулой бернулли:  $P(k=70)= C_{144}^{70}0.5^{70}0.5^{144-70} = \frac{144!}{70!(74)!}0.5^{70}0.5^{74} \approx 0.063 $

```python
print(f'Точный ответ к заданию 3 => {combinations(144,70) * 0.5**70 * 0.5**74}')
```

## Задание 4  
В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. 
Из каждого ящика вытаскивают случайным образом по два мяча.  
>а) Какова вероятность того, что все мячи белые?  
б) Какова вероятность того, что ровно два мяча белые?  
в) Какова вероятность того, что хотя бы один мяч белый?
### Решение  
а) По условию задачи из каждого ящика надо вытащить по два мяча. Верятность вытащить два мяча из каждого ящика равна числу случаев, благоприятствующим этому событию, к общему числу случаев =>  
Для первого ящика $ P_1 = \frac{C_7^2}{C_{10}^2} = \frac{\frac{7!}{5!2!}}{\frac{10!}{8!2!}} = \frac{6*7}{9*10} = \frac{42}{90}$  
Для второго ящика $ P_2 = \frac{C_9^2}{C_{11}^2} = \frac{\frac{9!}{7!2!}}{\frac{11!}{9!2!}} = \frac{8*9}{10*11} = \frac{72}{110}$  
Чтобы все мячи были белыми, необходимо по два белых белых мяча вытащить из каждой корзины => $P = P_1 * P_2 = \frac{42*72}{90*110} \approx 0.305$  
Вариант 2 - решение по формуле полной вероятности:  $P = \frac{7}{10} \frac{6}{90} \frac{9}{11} \frac{8}{10} = \frac{42*72}{90*110} \approx 0.305$  
б) Событию, что два мяча белые сопутствуют следующие исходы:  
1 - Из первого ящика вытащили два шара, а из второго ноль шаров. Вытаскивание ноля белых шаров из второго ящика равнозначно вытаскиванию двух черных шаров, то есть числу сочетаний два из двух.

$ P_{20} = \frac{C_7^2}{C_{10}^2} \frac{C_2^2}{C_{11}^2}= \frac{\frac{7!}{5!2!}}{\frac{10!}{8!2!}} \frac{\frac{2!}{0!2!}}{\frac{11!}{9!2!}} = \frac{6*7}{9*10} \frac{2}{10*11} = \frac{42}{90}\frac{1}{55}= \frac{42}{4950} \approx 0.0085 $

2 - Из второго ящика вытащили два шара, а из первого ноль шаров.   
Аналогично предыдущему случаю  
$ P_{02} = \frac{C_3^2}{C_{10}^2} \frac{C_9^2}{C_{11}^2} = \frac{\frac{3!}{1!2!}}{\frac{10!}{8!2!}}  \frac{\frac{9!}{7!2!}}{\frac{11!}{9!2!}} = \frac{3}{45} \frac{72}{110} = \frac{1}{15}\frac{72}{110}= \approx 0.0436 $  
3 - Из первого ящика вытащили один шар и из второго ящика вытащили один шар.  
По прежнему вероятность будем считать отношением благоприятных исходов к общему числу исходов.  
$ P_{11} = \frac{C_7^1 С_3^1}{C_{10}^2} \frac{C_9^1 С_2^1}{C_{11}^2}  = \frac{\frac{7!3!}{6!1!2!1!}}{\frac{10!}{2!8!}} \frac{\frac{9!2!}{8!1!1!1!}}{\frac{11!}{2!9!}}=\frac{7*3}{45}\frac{9*2}{55} \approx 0.1527$  
$P=P_{20}+P_{02}+P_{11} \approx 0.2048 $  
в) Посчитаем вероятность, когда все шары черные:  
$P_b=\frac{C_3^2}{C_{10}^2} \frac{C_2^2}{C_{11}^2} = \frac{3}{45} \frac{1}{55}=\frac{1}{825}$  
Сумма вероятностей, что все шары черные и хотя бы один шар белый равна 1 =>  
тогда $P=1-\frac{1}{825} = \frac{824}{825}$
