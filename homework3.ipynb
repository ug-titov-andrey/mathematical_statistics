```python
from math import factorial
from  math import exp
def combinations(n,k):
    return factorial(n)/(factorial(n-k)* factorial(k))
def poison(l,m):
    return (exp(-l)*(l**m))/factorial(m)
```

## Задание 1  
> Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 
Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, 
среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.  
### Решение  
Среднее арифметическое:  $\overline{x} = \frac{\sum\limits_{i=1}^{20}{x_i}}{20}=\frac{100+80+........+90+150}{20} = 65.3$  
Смещенная дисперсия: $ s^2=\frac{\sum\limits_{i=1}^{n}{(x-\overline{x})^2}}{n} = \frac{\sum\limits_{i=1}^{20}{(x-\overline{x})^2}}{20} = 950.11 $  
Несмещенная дисперсия: $ s^2=\frac{\sum\limits_{i=1}^{n}{(x-\overline{x})^2}}{n-1} = \frac{\sum\limits_{i=1}^{20}{(x-\overline{x})^2}}{19} \approx 1000.116 $  
Среднее квадратическое отклонение: $ s = \sqrt{s^2} = \sqrt{1000.116} \approx  31.625$

```python
import math
import numpy as np
from functools import reduce
A = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
my_mean = (reduce(lambda x,y: x+y, A))/len(A)
s_kv =  (reduce(lambda x,y: x+y, (A-my_mean)**2))/len(A)
s_kv_n =  (reduce(lambda x,y: x+y, (A-my_mean)**2))/(len(A)-1)
print(f'Среднее арифметическое => {my_mean}')
print(f'Смещенная дисперсия => {s_kv}')
print(f'Несмещенная дисперсия => {s_kv_n}')
print(f'Среднее квадратическое отклонение => {math.sqrt(s_kv_n)}')
```

$P_{0-3} = \frac{C_3^2}{C_8^2}\frac{C_5^3 C_7^1}{C_{12}^4}$  
$P = P_{2-1} + P_{1-2} +  P_{0-3} \approx 0.3687 $

$P(B_2) = \frac{0.7*0.25}{0.825} \approx 0.2424$  
$P(B_3) = \frac{0.9*0.5}{0.825} \approx 0.5454$

## Задание 4  
> Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна $P_1=0.1$, для второй - $P_2= 0.2$, для третьей - $P_3 = 0.25$. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?  
### Решение  
a)$P = P_1 * P_2 * P_3 = 0.1*0.2*0.25 = 0.005 $  
б)$P = (1-P_1) * P_2 * P_3 + P_1 * (1-P_2) * P_3  + P_1 * P_2 * (1-P_3) = 0.9*0.2*0.25 + 0.1*0.8*0.25 + 0.1*0.2*0.75 = 0.08$  
в) $P =1 -  (1-P_1) * (1-P_2) * (1-P_3) =1 -  0.9*0.8*0.75 = 0.46 $  
г) $P = 0.46 - 0.005 =0.455 $
